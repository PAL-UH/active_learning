```python
from skimage import io
import matplotlib.pyplot as plt
import cv2
import numpy as np 
%matplotlib inline
```

# Preprocessing

### Getting Triangles

Since we dont have enough sample, I filiped the images! 

```python
image3 = io.imread("triangles.PNG",as_grey=True)
image3.shape
```

```python
plt.imshow(image3)
plt.show()
```

```python
all_3 = [] 
width=200
for i in range(4):
    for j in range(4):
        crop = image3[(0+i*250):(270+i*260),(0+j*420):(420+j*420)]
        horizontal_img = crop.copy()
        vertical_img = crop.copy()
        both_img = crop.copy()
        horizontal_img = cv2.flip( crop, 0 )
        vertical_img = cv2.flip( crop, 1 )
        both_img = cv2.flip( crop, -1 )
        crop = cv2.resize(crop, (width, width))
        horizontal_img = cv2.resize(horizontal_img, (width, width))
        vertical_img = cv2.resize(vertical_img, (width, width))
        both_img = cv2.resize(both_img, (width, width))
        all_3.append(crop)
        all_3.append(horizontal_img)
        all_3.append(vertical_img)
        all_3.append(both_img)
```

```python
import random
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
n = len(all_3)
plt.figure(figsize=(20, 20))
for i in range(36):
    plt.subplot(6, 6, i+1)
    plt.imshow(all_3[i])
    
```

### Getting Rectangles 

Same process in here too!

```python
image4 = io.imread("squares.PNG",as_grey=True)
image4.shape
```

```python
plt.imshow(image4)
plt.show()
```

```python
all_4 = [] 
width=200
for i in range(4):
    for j in range(4):
        crop = image4[(0+i*250):(270+i*280),(0+j*420):(420+j*440)]
        horizontal_img = crop.copy()
        vertical_img = crop.copy()
        both_img = crop.copy()
        horizontal_img = cv2.flip( crop, 0 )
        vertical_img = cv2.flip( crop, 1 )
        both_img = cv2.flip( crop, -1 )
        crop = cv2.resize(crop, (width, width))
        horizontal_img = cv2.resize(horizontal_img, (width, width))
        vertical_img = cv2.resize(vertical_img, (width, width))
        both_img = cv2.resize(both_img, (width, width))
        all_4.append(crop)
        all_4.append(horizontal_img)
        all_4.append(vertical_img)
        all_4.append(both_img)
```

```python


n = len(all_4)
plt.figure(figsize=(20, 20))
for i in range(36):
    plt.subplot(6, 6, i+1)
    plt.imshow(all_4[i])
    
```

```python
train = []
labels = [] 
```

```python
for j in range(2):
    for i in range(len(all_3)): 
        train.append(all_3[i])
        labels.append(0)
    for i in range(len(all_4)): 
        train.append(all_4[i])
        labels.append(1)
```

```python
train = np.array(train)
labels = np.array(labels)
```

```python
train.shape
```

```python
train = train.reshape(len(train), 40000)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
```

```python
clf = LogisticRegression()
scores = cross_val_score(clf, train, labels, cv=6)
print(np.mean(scores))
```

```python
clf= LogisticRegression(C=100).fit(train,labels)
```

```python
test1 = io.imread("1.png", as_grey=True)
test1 = cv2.resize(test1, (200,200))
test_1 = test1.reshape(1, 40000)
pred1 = clf.predict(test_1)
plt.imshow(test1)
if pred1[0]==0:
    tit = "Triangle!"
else:
    tit = "Square!"
plt.title(tit)
plt.show()
```
